### Iris Flower Attributes
​
1. Sepal Length (cm)
*Description:* The length of the sepal, which is the outermost part of the flower that protects the petals.
​

​
2. Sepal Width (cm) 
*Description:* The width of the sepal, measured at its widest point.
​

​
3. Petal Length (cm)
*Description:* The length of the petal, which is the colorful inner part of the flower.
​

​
4. Petal Width (cm) 
*Description:* The width of the petal, measured at its widest point.
​

​
5. Species
*Description:* The species of iris flower, which can be one of three classes: Setosa, Versicolor, or Virginica.
​

```python
##kde Plot
numerical_features = [ col for col in df.columns if df[col].dtype in ['int64','float64']]

fig, axes = plt.subplots(2,2,figsize=(12,6))
axes = axes.flatten()
for i,col in enumerate(numerical_features): 
    sns.kdeplot(data=df,x=col, hue=df['Species'], shade=True, ax=axes[i])
    axes[i].set_title(col)
    
plt.tight_layout()
plt.show()
```

```python
##Scatter Plot
# scatter plot of petal length by petal width
sns.scatterplot(df,x=df['PetalLengthCm'],y=df['PetalWidthCm'],hue='Species')
```

```python
# scatter plot of sepal length by sepal width
sns.scatterplot(df,x=df['SepalLengthCm'],y=df['SepalWidthCm'],hue='Species')
```

```python
##Pair Plot
sns.pairplot(df,hue='Species')
```

```python
##histogram
numeric_features = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]

# Set up the figure and axes for subplots
fig, axes = plt.subplots(2,2,figsize=(12,6))

axes = axes.flatten()

for i, col in enumerate(numeric_features):
    sns.histplot(df[col], kde=True,edgecolor='black', ax=axes[i],)
    axes[i].set_title(col)
    axes[i].set_facecolor('lightpink')  # background color

plt.tight_layout()
plt.show()
```
